```python
africancountries = df_africancountries['Country'].values.tolist()
```

```python
#Extract records for respondents with country outside Africa
df_survey_africa = df_survey[df_survey.Country.isin(africancountries)]
df_survey_africa.reset_index(drop=True)
```

```python
#count of respondents who reported a country within Africa
total_respondents_africa = df_survey_africa['Respondent'].count()
```

```python
print("01. Percentage of Respondents who listed an African country is {:.1%}".format(total_respondents_africa/total_respondents) )
```

```python
#Grouping Reacords at Country Level with Count of Respondents
df_survey_africa_country = df_survey_africa.groupby(['Country'])['Respondent'].count().reset_index()
df_survey_africa_country['Percentage_Contribution'] = (df_survey_africa_country['Respondent']/total_respondents_africa)
df_survey_africa_country.sort_values(by='Percentage_Contribution', ascending=False)
```

```python
per_contribution_top3 = df_survey_africa_country.nlargest(4, columns='Percentage_Contribution').Percentage_Contribution.sum()
print("02. Within Africa, the top 4 countries with over {:.1%} of respondents are English speaking".format(per_contribution_top3))
```

```python
Top4_Africa_Countries = df_survey_africa[df_survey_africa['Country'].isin(['South Africa','Nigeria','Egypt','Kenya'])]
Top4_Africa_Countries
```

```python
Top4_Africa_Countries_Edu = Top4_Africa_Countries.groupby(['FormalEducation'])['Respondent'].count().reset_index()
Top4_Africa_Countries_Edu['Percentage_Contribution'] = (Top4_Africa_Countries_Edu['Respondent']/Top4_Africa_Countries_Edu['Respondent'].sum())
Top4_Africa_Countries_Edu.sort_values(by='Respondent', ascending=False)
```
